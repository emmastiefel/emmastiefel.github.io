```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import numpy as np
```

```python
all_albums_url = 'https://doolsetbangtan.wordpress.com/bts-albums/'
all_albums_request = requests.get(all_albums_url)
```

```python
##collect all album links
all_albums_content = all_albums_request.content
all_albums_soup = BeautifulSoup(all_albums_content, "html.parser")

raw_album_links = all_albums_soup.find(class_='entry-content').findAll('p')
album_links = []
for i in raw_album_links:
    try:
        link = i.find('a', href=True)
        album_links.append(link['href'])
    except:
        pass
```

```python
##repeat process for solo projects/mixtapes
mixtapes_link = 'https://doolsetbangtan.wordpress.com/solo-projects/'
mixtapes_request = requests.get(mixtapes_link)

##collect all album links
mixtapes_content = mixtapes_request.content
mixtapes_soup = BeautifulSoup(mixtapes_content, "html.parser")

raw_mixtape_links = mixtapes_soup.find(class_='entry-content').findAll('p')
#add mixtapes to album links
for i in raw_mixtape_links:
    try:
        link = i.find('a', href=True)
        album_links.append(link['href'])
    except:
        pass
```

```python
##repeat process for solo projects/mixtapes
mixtapes_link = 'https://doolsetbangtan.wordpress.com/solo-projects/'
mixtapes_request = requests.get(mixtapes_link)

##collect all album links
mixtapes_content = mixtapes_request.content
mixtapes_soup = BeautifulSoup(mixtapes_content, "html.parser")

raw_mixtape_links = mixtapes_soup.find(class_='entry-content').findAll('p')
#add mixtapes to album links
for i in raw_mixtape_links:
    try:
        link = i.find('a', href=True)
        album_links.append(link['href'])
    except:
        pass
```

```python
print(album_links)
```

```python
##access album content and store
album_soups = []
for album in album_links:
    this_request = requests.get(album)
    this_content = this_request.content 
    this_soup = BeautifulSoup(this_content, "html.parser")
    
    album_soups.append(this_soup)
    time.sleep(10) #sleep to be nice
```

```python
song_data = pd.DataFrame(columns=['Title', 'Album', 'URL', 'Soup'])

for soup in album_soups:
    raw_links = soup.find(class_='entry-content').findAll('p')
    #skip if link doesnt have href
    for i in raw_links:
        try:
            link = i.find('a', href=True)
            url = link['href']
        except:
            continue

        #only take internal links for song translations
        if 'doolsetbangtan' in url:
            print(link['href'])
            #start storing album, url, title info 
            song_name = link.get_text()
            song_album = soup.find(class_='entry-title').get_text()
            song_row = pd.DataFrame(data=[[song_name, song_album, url]], columns=['Title', 'Album', 'URL'])
            song_data = pd.concat([song_data, song_row], ignore_index=True)
```

```python
##add miscelllaneous collaborations and soundcloud songs
soundcloud_url = 'https://doolsetbangtan.wordpress.com/soundcloud/'
miscelllaneous_url = 'https://doolsetbangtan.wordpress.com/miscellaneous/'
soundcloud_request = requests.get(soundcloud_url)
misc_request = requests.get(miscelllaneous_url)

##collect all soups
soundcloud_content = soundcloud_request.content
misc_content = misc_request.content

soundcloud_soup = BeautifulSoup(soundcloud_content, "html.parser")
misc_soup = BeautifulSoup(misc_content, "html.parser")


for soup in [soundcloud_soup, misc_soup]:
    raw_links = soup.find(class_='entry-content').findAll('p')
    #skip if link doesnt have href
    for i in raw_links:
        try:
            link = i.find('a', href=True)
            url = link['href']
        except:
            continue

        #only take internal links for song translations
        if 'doolsetbangtan' in url:
            print(link['href'])
            #start storing album, url, title info 
            song_name = link.get_text()
            song_album = soup.find(class_='entry-title').get_text()
            song_row = pd.DataFrame(data=[[song_name, song_album, url]], columns=['Title', 'Album', 'URL'])
            song_data = pd.concat([song_data, song_row], ignore_index=True)
```

```python
##collect all song soups
for i, row in song_data.iterrows():
    url = row["URL"]
    this_request = requests.get(url)
    this_content = this_request.content 
    this_soup = BeautifulSoup(this_content, "html.parser")
    
    #store in dataframe
    song_data.at[i, 'Soup'] = this_soup

    print(i, url)
    time.sleep(10) #sleep to be nice
```

```python
lyrics_data = song_data.copy()
lyrics_data.insert(2, "English Lyrics Only", ['' for i in range(len(lyrics_data))])
lyrics_data.insert(2, "Translated Korean Lyrics Only", ['' for i in range(len(lyrics_data))])
lyrics_data.insert(2, "Notes", ['' for i in range(len(lyrics_data))])
lyrics_data.insert(2, "Complete Translated Lyrics", ['' for i in range(len(lyrics_data))])
lyrics_data.head()
```

```python
data = lyrics_data
data.insert(6, "Internal Links", ['' for i in range(len(data))])
```

```python
for i, row in data.iterrows():

    #get html contents for this song
    this_soup = row["Soup"]

    #get title:
    title = this_soup.find(class_='entry-header').find('h1').get_text()
    data.at[i, 'Title'] = title

    english_spans = this_soup.find(class_='entry-content').findAll('span', attrs={"style": "color:#800000;"})
    other_color_english = this_soup.find(class_='entry-content').findAll('span', attrs={"style": "color:#993300;"}) #another color for translations used in 4 oclock at least
    english_lyrics = " \n".join([s.get_text() for s in english_spans]) #all song lyrics in english (original or translated)
    other_color = " \n".join([s.get_text() for s in other_color_english]) 
    english_lyrics = english_lyrics + other_color

    notes_spans = this_soup.find(class_='entry-content').findAll('span', attrs={"style": "color:#008000;"})
    notes_lyrics = " \n".join([s.get_text() for s in notes_spans]) #all notes coded as green text

    english_original = this_soup.find(class_='entry-content').findAll('em')
    english_original_lyrics = " \n".join([s.get_text() for s in english_original]) #all song lyrics originally in english, not translated

    #get lyrics that were translated from korean (remove original )
    translated_lyrics = []
    for s in english_spans:
        translated = s.get_text()
        #rremove words originally in english
        for e in s.findAll('em'):
            translated = translated.replace(e.get_text(), '', 1)
        translated_lyrics.append(translated)
    translated_lyrics = " \n".join([s for s in translated_lyrics])

    #remove notes that mightve been mistakenly included in lyrics
    for note in notes_lyrics.split(' \n'):
        #only remove notes that contain at least two characters to avoid random tiny spans that mess things up (?)
        if len(note) >= 2:
            english_lyrics = english_lyrics.replace(note, '', 1)
            translated_lyrics = translated_lyrics.replace(note, '', 1)

    ##find all links to other doolset translations on this page
    #presumably, these are other songs connected to that song!! it looks like she consistently linked them
    all_as = this_soup.find(class_='entry-content').findAll('a', href=True)
    #check if link is an internal linke to a song (indicated by the /20, since all translations are apparently dated blog posts)
    internal_links = []
    for a in all_as:
        #filter out links to the same song and social media/comments pages
        if 'doolsetbangtan.wordpress.com/20' in a['href'] and row["URL"] not in a['href'] and '?' not in a['href']:
            internal_links.append(a['href'])

    data.at[i, 'Complete Translated Lyrics'] = english_lyrics
    data.at[i, 'Notes'] = notes_lyrics
    data.at[i, 'Translated Korean Lyrics Only'] = translated_lyrics
    data.at[i, 'English Lyrics Only'] = english_original_lyrics
    data.at[i, 'Internal Links'] = internal_links
```

```python
#data = data.drop_duplicates(subset=['Title']) #rremove songs duplicated in other albums
data.to_csv('doolset-lyrics-v6.csv')
```
