```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import numpy as np
```

```python
data = pd.read_csv('doolset-lyrics-v6.csv')
data.head()
```

```python
data.insert(10, "Youtube Video", ['' for i in range(len(data))])
```

```python
for i, row in data.iterrows():
    this_soup = row["Soup"]
    this_soup = BeautifulSoup(this_soup, "html.parser")

    yt_embed = this_soup.find(class_='jetpack-video-wrapper')
    if yt_embed:
        iframe = yt_embed.find('iframe', src=True)

        ##scrape youtube links from embeds
        request = requests.get(iframe['src'])
        content = request.content
        soup = BeautifulSoup(content, "html.parser")

        #get yt link from tthe embed (apparently its the only link idk)
        yt_link = soup.find('a', href=True)
        yt_links.append(yt_link['href'])
        print(yt_link['href'])

        data.at[i, "Youtube Video"] = yt_link['href']
        time.sleep(5)
```

```python
data.insert(10, "Youtube Views", [0 for i in range(len(data))])
```

```python
for i, row in data.iterrows():
    link = row["Youtube Video"]

    if link != '':
        ##access yt page
        request = requests.get(link)
        content = request.content
        soup = BeautifulSoup(content, "html.parser")

        #scrape views from page
        try:
            views = soup.select_one('meta[itemprop="interactionCount"][content]')['content']
            print(views)
            data.at[i, "Youtube Views"] = views
        except:
            print(link)
        time.sleep(5)
```
