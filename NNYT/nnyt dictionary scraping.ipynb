```python
import requests
from bs4 import BeautifulSoup, Comment
import time
import pandas as pd
```

```python
##import csv of first words list
words_data = pd.read_csv('nnyt-data-v1.csv')
words_data.insert(9, "is_in_dictionary", [False for i in range(len(words_data))])
words_data.head()
```

```python
base_url = 'https://www.dictionary.com/browse/'
c = 0
for i, row in words_data.iterrows():
    c = c + 1
    print(i, c)
    if c > 700:
        #confirm it hasn't been filled
        if row["is_in_dictionary"] == False:
            #try to get ud page for this term
            term = row["full_text"]

            try_d = requests.get(base_url + term)
            time.sleep(6) #sleep to be nice to wiki servers

            #confirm search was successful
            if try_d.status_code == 200:
                this_soup = BeautifulSoup(try_d.content, 'html.parser') 
                #get  header and lowercase it
                d_title = this_soup.find("h1").get_text().lower()

                #get and store relevant content if wiki header matches term
                if d_title == term:
                    words_data.at[i, "is_in_dictionary"] = True
                    print(term, "in dictionary.com")
words_data.to_csv('nnyt-data-v5-dictionary.csv')
```
