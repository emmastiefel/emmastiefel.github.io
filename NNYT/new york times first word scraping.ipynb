```python
import matplotlib.pyplot as plt
import numpy as np
import tweepy
import pandas as pd
import json
import datetime as dt
import time
import scipy
```

```python
###keys + getting twitter access###
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
bot_id = 'NYT_first_said'
nyt_first_said_bot = api.get_user(id=bot_id)
```

```python
#initialize dataframe to store all desired information
df = pd.DataFrame(columns=["id", #tweet id/link
                                    "full_text", #text of tweet (the wod)
                                    "favorite_count", #number of likes
                                    "retweet_count", #number of retweents
                                    "created_at"]) #date tweet was made/word was published
```

```python
#test with initial simple retrieval of tweets
initial_tweets = api.user_timeline(id=bot_id)
```

```python
#store info for initial 20 tweets in dataframe
for t in initial_tweets:
    print(t.text)
    df = df.append({"id": t.id_str, "full_text": t.text, "favorite_count":t.favorite_count, "retweet_count": t.retweet_count, "created_at": t.created_at}, ignore_index=True)
    oldest_id = t.id_str
    oldest_date = t.created_at
```

```python
#the most recent day we have archived data for
newest_archived = dt.datetime(2020, 4, 19)
last_text = ''

#iteratively try to retrieve tweets older than current oldest id 
while oldest_date > newest_archived:
    #attempt to retrieve and log statuses
    try:
        these_tweets = api.user_timeline(id=bot_id, max_id=oldest_id)
        #store tweets in dataframe
        for t in these_tweets:
            if t.text == last_text:
                print("ERROR REPEAT")
                continue
            print(t.text)
            
            df = df.append({"id": t.id_str, "full_text": t.text, "favorite_count":t.favorite_count, "retweet_count": t.retweet_count, "created_at": t.created_at}, ignore_index=True)

            oldest_id = t.id_str
            oldest_date = t.created_at 
            last_text = t.text
    #handle rate limits, wait 15 min
    except: 
        for i in range(15):
            print("wait ", 15 - i, " minutes") 
            time.sleep(1 * 60)
```

```python
#pull out all tweets longer than 1 word w/boolean mask and drrop
mask = [len(t) > 1 for t in df["full_text"].str.split()]
drop_index = df[mask].index
df.drop(labels=drop_index, inplace=True)
#confirm average length ======= 1
np.mean([len(i) for i in df["full_text"].str.split()])
```

```python
##write data frame to csv
df.to_csv('recent-nyt-first-said-data.csv')
```
