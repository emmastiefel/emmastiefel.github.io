```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from googletrans import Translator
import time
```

```python
##import csv of first words list
words_data = pd.read_csv('nnyt-data-v1.csv')
words_data.insert(9, "is_other_language", [False for i in range(len(words_data))])
words_data.insert(10, "guessed_language", [False for i in range(len(words_data))])

words_data.head()
```

```python
for i, row in words_data.iterrows():
    print(i)
    if i > 118:
        try:
            #attempt to translate the word
            translator = Translator()
            this_trans = translator.detect(row["full_text"])

            #only store words as a foreign language if google is 100% positive it is not english
            if this_trans.lang != 'en' and this_trans.confidence >= 1:
                words_data.at[i, "is_other_language"] = True
                words_data.at[i, "guessed_language"] = this_trans.lang
                print(this_trans.lang, this_trans.confidence, row["full_text"])
        except:
            print("wait 1 min")
            time.sleep(60)
```
