```python
import requests
from bs4 import BeautifulSoup, Comment
import time
import pandas as pd
```

```python
##import csv of first words list
words_data = pd.read_csv('nnyt-data-v1.csv', index_col=0)
words_data.insert(9, "is_in_wiki", [False for i in range(len(words_data))])
words_data.insert(10, "wiki_description", [False for i in range(len(words_data))])
words_data.head()
```

```python
base_url = 'https://en.wikipedia.org/wiki/'
c = 0
for i, row in words_data.iterrows():
    c = c + 1
    print(i, c)
    if c > 2000:
        #confirm it hasn't been filled
        if row["is_in_wiki"] == False:
            #try to get wiki page for this term
            term = row["full_text"]
            try_wiki = requests.get(base_url+term)
            time.sleep(6) #sleep to be nice to wiki servers

            #confirm search was successful
            if try_wiki.status_code == 200:
                this_soup = BeautifulSoup(try_wiki.content, 'html.parser') 
                #get wiki header and lowercase it
                wiki_title = this_soup.find("h1").get_text().lower()

                #get and store relevant content if wiki header matches term
                if wiki_title == term:
                    wiki_description = this_soup.find("p").get_text()
                    words_data.at[i, "is_in_wiki"] = True
                    words_data.at[i, "wiki_description"] = str(wiki_description)
                    print(term, "in wikipedia")
words_data.to_csv('nnyt-data-v3-wiki.csv')
```
