```python
import requests
import time
import pandas as pd
```

```python
##import csv of first words list
words_data = pd.read_csv('nnyt-data-v1.csv', index_col=0)
words_data.insert(9, "article_search_json", [False for i in range(len(words_data))])
words_data.head()
```

```python
##store raw json of first article search result for every word, date pair

overall_url = 'https://api.nytimes.com/svc/search/v2/articlesearch.json'

for i, row in words_data.iterrows():
    print(i)
    #only run if thing hasn't been filled in
    if row["article_search_json"] == False and i >= 2790:
        word = row["full_text"]
        year = 'pub_year:'+str(row["date"][:4])+'' 

        this_parameters = {
            "q": word,
            'fq': year,
            "api-key": ''
        }

        #get api results
        this_result = requests.get(overall_url, params=this_parameters)
        #sleep to respect per-minute rate limits
        time.sleep(6)

        #if there's at least one hit, store it :)
        if this_result.json()['response']['meta']['hits'] >= 1:
            words_data.at[i, "article_search_json"] = str(this_result.json()['response']['docs'][0])
            print(this_result.json()['response']['docs'][0]['web_url'])
words_data.to_csv('nnyt-data-v2-article-search.csv')
```
